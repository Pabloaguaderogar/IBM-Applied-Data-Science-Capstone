## Tasks

Now write and execute SQL queries to solve the assignment tasks.

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
%%sql
SELECT DISTINCT "Launch site" 
FROM SPACEXTABLE;
```

```python
%%sql
SELECT DISTINCT "Customer" FROM SPACEXTABLE
WHERE "Customer" Like "%NASA%";
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
%%sql
SELECT * FROM SPACEXTABLE 
WHERE "Launch site" LIKE 'CCA%' 
LIMIT 5;
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
%%sql
SELECT SUM("Payload mass") AS "Total_Payload_Mass_NASA_CRS"
FROM SPACEXTABLE
WHERE Customer Like '%NASA%';
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
%%sql
SELECT AVG(CAST(REPLACE(REPLACE("Payload mass", 'kg', ''), ',', '') AS FLOAT)) AS "Average_Payload_Mass_v1.1"
FROM SPACEXTABLE
WHERE "Version Booster" LIKE 'F9 v1.1%';
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
%%sql
-- 1. Ver los nombres reales de las columnas en la tabla
-- (Por si se importó como "Landing_Outcome" o similar)
SELECT * FROM SPACEXTABLE LIMIT 1;

-- 2. Ver todos los valores únicos de la columna de resultados
-- (Para ver el texto exacto que guardó el scraping)
SELECT DISTINCT "Booster landing" FROM SPACEXTABLE;
```

```python
%%sql
SELECT MIN("Date") AS "First_Successful_Ground_Landing"
FROM SPACEXTABLE
WHERE LOWER("Booster landing") LIKE '%success%';
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
%%sql
SELECT DISTINCT "Version Booster"
FROM SPACEXTABLE
WHERE "Booster landing" = 'Success'
  AND CAST(REPLACE(REPLACE("Payload mass", 'kg', ''), ',', '') AS FLOAT) BETWEEN 4000 AND 6000;
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
%%sql
SELECT "Booster landing", COUNT(*) AS "Total_Count"
FROM SPACEXTABLE
WHERE "Booster landing" IN ('Success', 'Failure')
GROUP BY "Booster landing";
```

### Task 8



##### List all the booster_versions that have carried the maximum payload mass, using a subquery with a suitable aggregate function.


```python
%%sql
SELECT DISTINCT "Version Booster", "Payload mass"
FROM SPACEXTABLE
WHERE CAST(REPLACE(REPLACE("Payload mass", 'kg', ''), ',', '') AS FLOAT) = (
    SELECT MAX(CAST(REPLACE(REPLACE("Payload mass", 'kg', ''), ',', '') AS FLOAT))
    FROM SPACEXTABLE
);
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 6,2) as month to get the months and substr(Date,0,5)='2015' for year.**


```python
%%sql
SELECT "Date", "Booster landing", "Version Booster"
FROM SPACEXTABLE
WHERE "Date" LIKE '%2015%'
LIMIT 10;
```

```python
%%sql
SELECT 
    CASE 
        WHEN "Date" LIKE '%January%' THEN 'January'
        WHEN "Date" LIKE '%February%' THEN 'February'
        WHEN "Date" LIKE '%March%' THEN 'March'
        WHEN "Date" LIKE '%April%' THEN 'April'
        WHEN "Date" LIKE '%May%' THEN 'May'
        WHEN "Date" LIKE '%June%' THEN 'June'
        WHEN "Date" LIKE '%July%' THEN 'July'
        WHEN "Date" LIKE '%August%' THEN 'August'
        WHEN "Date" LIKE '%September%' THEN 'September'
        WHEN "Date" LIKE '%October%' THEN 'October'
        WHEN "Date" LIKE '%November%' THEN 'November'
        WHEN "Date" LIKE '%December%' THEN 'December'
    END AS "Month",
    "Booster landing" AS "Landing_Outcome",
    "Version Booster",
    "Launch site"
FROM SPACEXTABLE
WHERE "Booster landing" LIKE '%Failure%' 
  AND "Date" LIKE '%2015%';
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
%%sql
SELECT "Booster landing", COUNT(*) AS "Outcome_Count"
FROM (
    SELECT "Booster landing",
        -- Construimos una fecha comparable YYYYMMDD
        (SUBSTR("Date", -4) || 
         CASE 
            WHEN "Date" LIKE '%January%' THEN '01' WHEN "Date" LIKE '%February%' THEN '02'
            WHEN "Date" LIKE '%March%' THEN '03'   WHEN "Date" LIKE '%April%' THEN '04'
            WHEN "Date" LIKE '%May%' THEN '05'     WHEN "Date" LIKE '%June%' THEN '06'
            WHEN "Date" LIKE '%July%' THEN '07'    WHEN "Date" LIKE '%August%' THEN '08'
            WHEN "Date" LIKE '%September%' THEN '09' WHEN "Date" LIKE '%October%' THEN '10'
            WHEN "Date" LIKE '%November%' THEN '11' WHEN "Date" LIKE '%December%' THEN '12'
         END || 
         CASE 
            WHEN LENGTH(SUBSTR("Date", 1, INSTR("Date", ' ') - 1)) = 1 
            THEN '0' || SUBSTR("Date", 1, INSTR("Date", ' ') - 1)
            ELSE SUBSTR("Date", 1, INSTR("Date", ' ') - 1)
         END) AS "Full_Date_Numeric"
    FROM SPACEXTABLE
)
WHERE "Full_Date_Numeric" BETWEEN '20100604' AND '20170320'
GROUP BY "Booster landing"
ORDER BY "Outcome_Count" DESC;
```
